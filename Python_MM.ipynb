```python
#Read in gene expression matrices for X and Y and the two files 
#containing information on the differentially expressed genes

GEM_X_file = open('GEM_X.csv', 'r')
GEM_Y_file = open('GEM_Y.csv', 'r')
variety_X_file = open('Leaf_DEGs_VarX_T1.csv', 'r')
variety_Y_file = open('Leaf_DEGs_VarY_T1.csv', 'r')

#empty lists to hold split lines of csv files
GEMX_split = []
GEMY_split = []
VARY_split = []
VARX_split = []

#read lines of GEM X file
for line in GEM_X_file.readlines():
    line_split = line.split(',')
    GEMX_split.append(line_split)

#read lines of GEM Y file
for line in GEM_Y_file.readlines():
    line_split = line.split(',')
    GEMY_split.append(line_split)

#read lines of Variety X file
for line in variety_X_file.readlines():
    line_split = line.split(',')
    VARX_split.append(line_split)

#read lines of Variety Y file
for line in variety_Y_file.readlines():
    line_split = line.split(',')
    VARY_split.append(line_split)
```

```python
#strip all values of trailing white spaces - to avoid errors in R - ie things not matching

#rid of trailing white spaces for GEMX
for i in range(len(GEMX_split)):
    for i2 in range(len(GEMX_split[i])):
        GEMX_split[i][i2] == GEMX_split[i][i2].rstrip()
        
#rid of trailing white spaces for GEMY
for i in range(len(GEMY_split)):
    for i2 in range(len(GEMY_split[i])):
        GEMY_split[i][i2] == GEMY_split[i][i2].rstrip()

#rid of trailing white spaces for VARX
for i in range(len(VARX_split)):
    for i2 in range(len(VARX_split[i])):
        VARX_split[i][i2] == VARX_split[i][i2].rstrip()
        
#rid of trailing white spaces for VARY
for i in range(len(VARY_split)):
    for i2 in range(len(VARY_split[i])):
        VARY_split[i][i2] == VARY_split[i][i2].rstrip()
```

```python
#Create ouput: Two gene expression matrices (one for each Variety), 
#each with a header, and containing gene expression from two conditions only: 
#the control condition (in columns 2-4; replicates 1-3) 
#and stress treatment condition 1 (columns 5-7; replicates 1-3). 

#empty lists to hold reformated gene expression matrices
gemX = []
gemY = []
leafX = []
leafY = []

#first append the headers to each list in the order wanted
gemXheaders = [GEMX_split[0][0],GEMX_split[0][1],GEMX_split[0][4],GEMX_split[0][8],GEMX_split[0][2],GEMX_split[0][6],GEMX_split[0][10]+'\n']
gemX.append(gemXheaders)

gemYheaders = [GEMY_split[0][0],GEMY_split[0][1],GEMY_split[0][5],GEMY_split[0][9],GEMY_split[0][3],GEMY_split[0][7],GEMY_split[0][11]]
gemY.append(gemYheaders)

leafXheaders = ['gene_name',VARX_split[0][2],VARX_split[0][6][:-1],VARX_split[0][4],VARX_split[0][5],GEMX_split[0][1],GEMX_split[0][4],GEMX_split[0][8],GEMX_split[0][2],GEMX_split[0][6],GEMX_split[0][10]+'\n']
leafX.append(leafXheaders)

leafYheaders = ['gene_name',VARY_split[0][2],VARY_split[0][6][:-1],VARY_split[0][4],VARY_split[0][5],GEMY_split[0][1],GEMY_split[0][5],GEMY_split[0][9],GEMY_split[0][3],GEMY_split[0][7],GEMY_split[0][11]]
leafY.append(leafYheaders)

#sort data into newly organized rows to append to each of gemX, gemY, leafX, leafY 
for i in range(1,len(GEMX_split)):
    newX_row = [GEMX_split[i][0],GEMX_split[i][1],GEMX_split[i][4],GEMX_split[i][8],GEMX_split[i][2],GEMX_split[i][6],GEMX_split[i][10]+'\n']
    gemX.append(newX_row)
    
for i in range(1,len(GEMY_split)):
    newY_row = [GEMY_split[i][0],GEMY_split[i][1],GEMY_split[i][5],GEMY_split[i][9],GEMY_split[i][3],GEMY_split[i][7],GEMY_split[i][11]]
    gemY.append(newY_row)
    
#intialize index variable
index = -1

#for each gene in VARX, iterating over gemX to find the matching gene
for i in range(1, len(VARX_split)):
    
    gene_to_find = VARX_split[i][0]
    
    for x, gene in enumerate(gemX):
        if gene[0] == gene_to_find:
            index = x
    
    #once matching index is found, can access replicate numbers and append new row to leafX
    leafX_row = [gene_to_find,VARX_split[i][2],VARX_split[i][6],VARX_split[i][4],VARX_split[i][5],gemX[index][1],gemX[index][2],gemX[index][3],gemX[index][3],gemX[index][5],gemX[index][6]]
    leafX.append(leafX_row)

#Same steps but for Leaf Variety Y
#intialize index variable
indY = -1

#for each gene in VARY, iterating over gemY to find the matching gene
for i in range(1, len(VARY_split)):
    
    gene_to_find = VARY_split[i][0]
    
    for y, geneY in enumerate(gemY):
        if geneY[0] == gene_to_find:
            indY = y
    
    #once matching index is found, can access replicate numbers and append new row to leafY
    leafY_row = [gene_to_find,VARY_split[i][2],VARY_split[i][6],VARY_split[i][4],VARY_split[i][5],gemY[indY][1],gemY[indY][2],gemY[indY][3],gemY[indY][3],gemY[indY][5],gemY[indY][6]]
    leafY.append(leafY_row)
```

```python
#create csv file outputs for each of the outputs above

import csv

#list gemX and gemY are the two gene expression matrices containing gene expression for control and stress conditions

#gemX csv will be all_VarX_TwoTimePoints.csv
with open('all_VarX_TwoTimePoints.csv', mode='w', newline='') as csvfile:
    #create file to write with csv writer
    csv_wr = csv.writer(csvfile)
    #write each row from data_reform to csv file
    csv_wr.writerows(gemX)
    
#gemY csv will be all_VarY_TwoTimePoints.csv
with open('all_VarY_TwoTimePoints.csv', mode='w', newline='') as csvfile:
    #create file to write with csv writer
    csv_wr = csv.writer(csvfile)
    #write each row from data_reform to csv file
    csv_wr.writerows(gemY)
    
#leafX csv will be Leaf_DEGs_VarX.csv
with open('Leaf_DEGs_VarX.csv', mode='w', newline='') as csvfile:
    #create file to write with csv writer
    csv_wr = csv.writer(csvfile)
    #write each row from data_reform to csv file
    csv_wr.writerows(leafX)
    
#leafY csv will be Leaf_DEGs_VarY.csv
with open('Leaf_DEGs_VarY.csv', mode='w', newline='') as csvfile:
    #create file to write with csv writer
    csv_wr = csv.writer(csvfile)
    #write each row from data_reform to csv file
    csv_wr.writerows(leafY)
```
